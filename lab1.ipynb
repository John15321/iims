```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
ls = 0
rs = 1
f = lambda x: np.exp(-(x**2) / 2) / (np.sqrt(2 * 3.14))
x = np.random.uniform(ls, rs, n)
y = np.random.uniform(ls, rs, n)

M = 0.3  # scale factor
u1 = np.random.rand(10000) * 3  # uniform random samples scaled out
u2 = np.random.rand(10000)  # uniform random samples

(idx,) = np.where(u2 <= f(u1) / M)  # rejection criterion
v1 = u1[idx]
v2 = u2[idx]
plt.scatter(v1, v2)
```

```python
plt.scatter(x, y)
plt.scatter(x * (x >= 3), f(x))
```

```python
(idx,) = np.where(y <= f(x))  # rejection criterion
v = x[idx]
plt.scatter(idx, v)
```
